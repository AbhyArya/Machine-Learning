```python
housing_num = housing.drop('ocean_proximity', axis = 1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
num_pipeline = Pipeline([('imputer', SimpleImputer(strategy = 'median')), ("std_scaler", StandardScaler())])
```

```python
num_pipeline
```

```python
housing_num_tr = num_pipeline.fit_transform(housing_num)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
num_attributes = list(housing_num)
cat_attributes = ['ocean_proximity']
full_pipeline = ColumnTransformer([('num', num_pipeline, num_attributes), ('cat', OneHotEncoder(), cat_attributes)])
housing_prepared = full_pipeline.fit_transform(housing)
```

```python
housing_original = pd.read_csv("https://raw.githubusercontent.com/mohitgupta-omg/Kaggle-California-Housing-Prices/master/Data/housing.csv")
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(housing_original)
```

```python
housing_label = housing["median_house_value"].copy()
```

```python
from sklearn.linear_model import LinearRegression

len_leg = LinearRegression()
len_leg.fit(housing_prepared, housing_label)
```

```python
some_data = housing.iloc[0:5]
some_labels = housing_label[0:5]

some_data_prepare = full_pipeline.transform(some_data)
print("predicted data", len_leg.predict(some_data_prepare))
print("actual data", list(some_labels))
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np;

housing_prediction = len_leg.predict(housing_prepared)

len_mse = mean_squared_error(housing_label, housing_prediction)
len_rmse = np.sqrt(len_mse)
len_rmse
```

```python
from sklearn.tree import DecisionTreeRegressor

t_leg =  DecisionTreeRegressor()
t_leg.fit(housing_prepared, housing_label)

housing_prediction = t_leg.predict(housing_prepared)

t_mse = mean_squared_error(housing_label, housing_prediction)
t_rmse = np.sqrt(t_mse)
t_rmse
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(t_leg, housing_prepared, housing_label, scoring="neg_mean_squared_error", cv=10)
t_leg_rmse_score = np.sqrt(-score)
t_leg_rmse_score
```
