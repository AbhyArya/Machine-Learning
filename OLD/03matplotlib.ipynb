```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
```

```python
plt.plot(x,y)
plt.show()
```

```python
#default x-point
              #0 1 2
yy = np.array([1,6,4])
plt.plot(yy)
plt.show()
```

```python
#marker
plt.plot(x,y,marker='o')#D,*,x,
plt.show()
                                            
```

```python
#marker
plt.plot(x,y,'D')#D,*,x,
plt.title("test")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
#sub plot

plt.subplot(1,2,1)
plt.plot(x,y,'o')
plt.title("ok")

plt.subplot(1,2,2)
plt.plot(x,y,'o')
plt.title('ok2')

plt.suptitle("hello")

plt.show()
```

```python
#scatter

plt.scatter(x,y)
```

```python
#bar graph
plt.bar(x,y)
plt.barh(x,y)
```

```python
#histogram
plt.hist(x)
```

```python
#pie chart
plt.pie(x,explode=[.2,0,0,0],shadow=True)
plt.show()
```
