```python
import pandas as pd
```

# Series
### Homogeneous


```python
my_list = [1, 2, 3,'series 1','series 2','series 3','series 4','series 5']
print(my_list)
pd_series = pd.Series(my_list, index = ["first", "second", "third", "fourth","five","six","seven","eight"])
print(pd_series)
```

```python
pd_series = pd.Series(4) # scaler series
print(pd_series)
pd_series = pd.Series(4, index = [_ for _ in range(10)]) # multiple item with same value 
print(pd_series)
```

```python
pd_series = pd.Series({"a":1,'b':2})
print(pd_series)
```

# Data Frame
### Heterogeneous

```python
pr_df = pd.DataFrame([[_ for _ in range(5)] for _ in range(10)], columns = ["first","second","third","fourth","fifth"])
pr_df
```

```python
pd_df = pd.DataFrame({"first":[_*_ for _ in range(5)],"second":[_**_ for _ in range(5)]})
pd_df
```

```python
myList = [[1,"Abhishek"],[2,"Vivek"],[3,"Ankit"],[4,"Pooja"]]
pd_df = pd.DataFrame([{"id":id,"name":name} for id, name in myList])
pd_df
```

```python
pd_df = pd.DataFrame({"id":pd.Series([1,2,3]),"name":pd.Series(["a",'b','c'])})
pd_df
```

```python
pd_df = pd.DataFrame([(1,'a'),(2,'b'),(3,'c')], columns = ['id','name'])
pd_df
```

# Merge Function

```python
df1 = pd.DataFrame({"Id":[1,2,3,4,5,6],'Class':[9,8,7,6,5,4]})
df2 = pd.DataFrame({"Id":[1,2,3,4,5,6],'Name':list("ABCDEF")})
print(df1)
print(df2)
pd.merge(df1,df2,"inner","Id")
# pd.merge?
```

# Concat Function

```python
pd.concat([df1,df2],axis = 1)
```

# Join Function


```python
df1.join(df2,rsuffix='_')
```
