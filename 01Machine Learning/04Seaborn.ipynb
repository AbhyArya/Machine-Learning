```python
import matplotlib.pyplot as plt
import matplotlib.style as style
import matplotlib.image as mpimg
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats 
```

# Lineplot

```python
df = sns.load_dataset('titanic')
df = df[df["fare"]<100]
df.head()
```

```python
# plt.figure(figsize = (16,9))
sns.lineplot(x = 'age',y = 'fare', data = df,style = 'sex',hue = 'sex', palette = 'hot',markers = ['o','<',])
plt.show()
```

# Histogram or distplot

```python
plt.figure(figsize = (16,9))
sns.set()
sns.histplot(df['fare'],kde = True,stat = 'density',color = 'k', line_kws = {"linewidth":8})
mean, std = stats.norm.fit(df['fare'])

dataPoint_array = np.linspace(df['fare'].min(), df['fare'].max(), 100)
pdf = stats.norm.pdf(dataPoint_array, mean, std)
plt.plot(dataPoint_array, pdf, label="Normal", color="red",linewidth = 8)
plt.legend()
plt.show()
```

# Bar plot

```python
plt.figure(figsize = (50,20))
sns.barplot(data = df,x = 'age', y = 'fare', ci = None)
plt.show()
```

# Scatter Plot

```python
sns.scatterplot(x='age',y='fare',data = df,hue = 'who',style = 'alive')
plt.show()
```

```python
plt.figure(figsize = (20,9))
sns.scatterplot(x='age',y='fare',data = df,hue = 'who',style = 'sex',size = 'who',sizes=(400,300))
plt.legend(fontsize = 25)
plt.show()
```

# Heatmap plot

```python
gw_df = pd.read_csv("C:\\Users\\Abhi Arya\\Downloads\\Who_is_responsible_for_global_warming.csv")
gw_df = gw_df.drop(columns = ["Country Code","Indicator Name","Indicator Code"],axis = 1).set_index("Country Name")
gw_df.head()
```

```python
plt.figure(figsize = (20,9))
sns.heatmap(data = gw_df, annot = True,linewidth = 5,linecolor="k")
plt.show()
```

```python
plt.figure(figsize = (20,9))
sns.heatmap(data = gw_df.corr(), annot = True,linewidth = 5,linecolor="k",mask = np.triu(gw_df.corr()) )
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights_df = flights.pivot("month", "year", "passengers")

plt.figure(figsize= (16,9)) 

cbar_kws = {"shrink":.8,
           'extend':'max',
           'extendfrac':.2, 
           "drawedges":True} 
 
sns.heatmap(flights_df.corr(), cmap="inferno", annot = True, linewidth = 2, cbar_kws=cbar_kws)
 
plt.title("Heatmap Correlation of 'Flights' Dataset", fontsize = 25)
plt.xlabel("Years", fontsize = 20)
plt.ylabel("Months", fontsize = 20)
plt.show()
```

```python
plt.figure(figsize= (16,9)) 

cbar_kws = {"shrink":.8,
           'extend':'both'} 

sns.heatmap(flights_df, cmap="inferno", annot = True, fmt = 'd', linewidth = 2, cbar_kws=cbar_kws)
 
plt.title("Heatmap of 'Flights' Dataset", fontsize = 25)
plt.xlabel("Years", fontsize = 20)
plt.ylabel("Months", fontsize = 20)
plt.show()
```

```python
# Cor-relation formula
# r = (sum((xi - mean(x))*(yi - mean(y))) / sqrt(sum(((xi - mean(x))**2)*((yi - mean(y))**2))

# r = correlation coefficient
# xi = values of the x-variable in a sample
# mean(x) =mean of the values of the x-variable
# yi = values of the y-variable in a sample
# mean(y) = mean of the values of the y-variable
```
