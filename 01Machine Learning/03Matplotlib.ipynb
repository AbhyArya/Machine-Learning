```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.style as style
import matplotlib.image as mpimg
```

# Line plot

```python
y1 = np.arange(10)
y2 = np.arange(10)**(1.4)
x = np.random.randint(50,size=(10))
print(x)
print(y1)
print(y2)
```

```python
style.use("ggplot")
plt.plot(x,y1,color='r',marker = 'o',linewidth = 3, markersize = 10,label='Motka')
plt.plot(x,y2,color='g',marker = 's',linewidth = 3, markersize = 10,label='Chudur_Budur')
plt.grid(color = 'b',linewidth = 2,alpha = .2)
plt.title("Ye to title h",color = 'b')
plt.xlabel("Kuch to x me h")
plt.ylabel("kuch to y me h")
plt.legend(loc = 2)
plt.axis([-10,50,-5,35])
plt.show()
```

# Histogram

```python
bins = [i for i in range(0,101,5)]
boys_size = np.random.randint(101,size = (100))
girl_size = np.random.randint(101,size = (100))
style.use("ggplot")
plt.hist([boys_size,girl_size],bins,color = ['m','k'],label=['Girls_vegina_size',"Boys_penis_size"])
plt.title("Size of Boys_penis and Girls_vegina")
plt.xlabel("Age of Boy/girl")
plt.ylabel("Size of Boys_penis/Girls_vegina")
plt.legend()
plt.show()
```

 # Bar Graph

```python
classes = ['bakwas','londiyabaazi','nikamapan','party']
classes_idx = np.arange(len(classes))
cse_1st_year = np.random.randint(1,100,size = (4));
cse_2st_year = np.random.randint(1,100,size = (4));
cse_3st_year = np.random.randint(1,100,size = (4));
cse_4st_year = np.random.randint(1,100,size = (4));
# print(cse_1st_year)
width = 0.2
style.use('ggplot')
# plt.figure(figsize = (20,9))
plt.bar(classes_idx,cse_1st_year,label = 'CSE 1st Year',width = 0.2)
plt.bar(classes_idx+width,cse_2st_year,label = 'CSE 2st Year',width = 0.2)
plt.bar(classes_idx+2*width,cse_3st_year,label = 'CSE 3st Year',width = 0.2)
plt.bar(classes_idx+3*width,cse_4st_year,label = 'CSE 4st Year',width = 0.2)
plt.xticks(classes_idx+width,classes)
plt.title("CSE Student work")
plt.ylabel("Amount of work")
plt.xlabel("Field of work")
plt.legend()
plt.axis([-.5,4,0,110])
plt.grid(color = 'g',alpha = 0.4,linewidth = 2)
plt.show()
```

# Scatter Plot

```python
df = pd.read_csv("C:\\Users\\Abhi Arya\\Downloads\\archive\\googleplaystore.csv",nrows = 1000)
df.shape
```

```python
df.fillna(method='ffill',inplace = True)
```

```python
rating = df['Rating']
review = df["Reviews"]
install = df["Installs"]
dataToGet = review<1e3
review = review[dataToGet]
rating = rating[dataToGet]
install = install[dataToGet]
```

```python
# plt.figure(figsize = (20,9))
plt.scatter(rating,review,marker='s',color = 'g',linewidth = 3,alpha = .5,label = 'reting-review')
plt.scatter(rating,install,marker='*',color = 'r',linewidth = 3,alpha = .3,label = 'reting-install')
plt.title("Google Play Store Rating and Reviews")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.legend()
plt.show()
```

# Pie Chart

```python
classes = ['bakwas','londiyabaazi','nikamapan','party']
classes_idx = np.arange(len(classes))
explode = [0,0,0.12,0.15]
cse_1st_year = np.random.randint(1,100,size = (4));
cse_2st_year = np.random.randint(1,100,size = (4));
cse_3st_year = np.random.randint(1,100,size = (4));
cse_4st_year = np.random.randint(1,100,size = (4));
plt.pie(cse_1st_year,labels = classes,shadow = True, explode = explode,autopct = "%0.1f%%",startangle = 90)
plt.title("Aliens Working",color = 'g')
plt.legend()
plt.show()
```

```python
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```
